### 4. Textual Data Preprocessing
(It's mandatory for textual dataset i.e., NLP, Sentiment Analysis, Text Clustering etc.)

#### 1. Expand Contraction

````python
!pip install contractions

import contractions

df['Customer Remarks'] = df['Customer Remarks'].astype(str).apply(contractions.fix)

df[['Customer Remarks']].head()
````

#### 2. Lower Casing

````python
df['Customer Remarks'] = df['Customer Remarks'].astype(str).str.lower()

df[['Customer Remarks']].head()
````

#### 3. Removing Punctuations

````python
import string

df['Customer Remarks'] = df['Customer Remarks'].astype(str).str.replace(
    f"[{string.punctuation}]",
    "",
    regex=True
)

df[['Customer Remarks']].head()
````

#### 4. Removing URLs & Removing words and digits contain digits.

````python
import re

df['Customer Remarks'] = df['Customer Remarks'].astype(str).apply(
    lambda x: re.sub(r'http\S+|www\S+|\w*\d\w*', '', x)
)

df[['Customer Remarks']].head()
````

#### 5. Removing Stopwords & Removing White spaces

````python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

df['Customer Remarks'] = df['Customer Remarks'].astype(str).apply(
    lambda x: " ".join(
        [word for word in x.split() if word.lower() not in stop_words]
    )
)

df[['Customer Remarks']].head()
````

````python
df['Customer Remarks'] = df['Customer Remarks'].astype(str).str.strip()

df[['Customer Remarks']].head()
````

#### 6. Rephrase Text

````python
# Rephrase Text

df['Customer Remarks'] = df['Customer Remarks'].astype(str)

# Display the processed text
df[['Customer Remarks']].head()
````

#### 7. Tokenization

````python
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('punkt_tab')

# Tokenization
df['Tokens'] = df['Customer Remarks'].astype(str).apply(word_tokenize)

df[['Customer Remarks', 'Tokens']].head()
````

#### 8. Text Normalization

````python
import nltk
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')
nltk.download('omw-1.4')

lemmatizer = WordNetLemmatizer()

df['Lemmatized_Text'] = df['Customer Remarks'].astype(str).apply(
    lambda x: " ".join([lemmatizer.lemmatize(word) for word in x.split()])
)

df[['Customer Remarks', 'Lemmatized_Text']].head()
````

##### Which text normalization technique have you used and why?

Text normalization was not required because this project does not involve text processing or NLP.

#### 9. Part of speech tagging

````python
import nltk
from nltk import pos_tag
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')

df['POS_Tags'] = df['Customer Remarks'].astype(str).apply(
    lambda x: pos_tag(word_tokenize(x))
)

df[['Customer Remarks', 'POS_Tags']].head()
````
